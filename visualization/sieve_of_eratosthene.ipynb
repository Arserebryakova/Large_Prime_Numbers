```python
import pandas as pd

def read_test_result(filename):
    data = pd.read_csv(f"../large_prime_numbers/test_results/{filename}")

    digits = data["Digits"].to_list()
    N_values = list(map(int, data["N"].to_list()))
    primes_found = list(map(int, data["found"].to_list()))
    durations = data["duration"].to_list()
    return digits, N_values, primes_found, durations
```

```python
import matplotlib.pyplot as plt
import numpy as np

#данные
digits, N_values, primes_found, durations = read_test_result("Sieve_23-09-2024 22-14-25.csv")


#график зависимости времени от количества цифр
plt.figure(figsize=(10, 6))
plt.scatter(digits, durations, color='b', label='Duration vs Digits')
plt.yscale('log')  
plt.xlabel('Digits')
plt.ylabel('Duration (seconds)')
plt.title('Sieve of Eratosthenes: Duration vs Number of Digits')
plt.grid(True)
plt.legend()
plt.show()


#график зависимости времени от значения N
plt.figure(figsize=(10, 6))
plt.scatter(N_values, durations, color='green', label='Duration vs N')
plt.yscale('log')  
plt.xscale('log')  
plt.xlabel('N')
plt.ylabel('Duration (seconds)')
plt.title('Sieve of Eratosthenes: Duration vs N')
plt.grid(True)

z = np.polyfit(np.log(N_values), np.log(durations), 1) 
p = np.poly1d(z)
x_trend = np.linspace(min(N_values), max(N_values), 100)
y_trend = np.exp(p(np.log(x_trend)))
plt.plot(x_trend, y_trend, color='green', label='Trend Line')

plt.legend()
plt.show()

#график зависимости времени от количества найденных чисел
plt.figure(figsize=(10, 6))
plt.scatter(primes_found, durations, color='blue', label='Duration vs Primes found')
plt.yscale('log')  
plt.xscale('log')  
plt.xlabel('Primes found')
plt.ylabel('Duration (seconds)')
plt.title('Sieve of Eratosthenes:  Duration vs Primes found')
plt.grid(True)

z = np.polyfit(np.log(primes_found), np.log(durations), 1)  
p = np.poly1d(z)
x_trend = np.linspace(min(primes_found), max(primes_found), 100)
y_trend = np.exp(p(np.log(x_trend)))
plt.plot(x_trend, y_trend, color='blue', label='Trend Line')


plt.legend()
plt.show()
```

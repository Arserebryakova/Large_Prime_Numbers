```python
import pandas as pd
import math
# Исходные данные CSV
data = pd.read_csv("../large_prime_numbers/test_results/Pollard_01-04-2025 21-28-11.csv")

# Группировка по 'number' и вычисление статистик:
stats = data.groupby('number')['duration'].agg(
    mean_duration='mean',           # среднее время
    median_duration='median',       # медианное время
).reset_index()

stats['number'] = stats['number'].astype(float)

stats['digits'] = stats['number'].apply(lambda x: len(str(int(x))))
stats['number_log10'] = stats['number'].apply(lambda x: math.log10(x))

print(stats)

data2 = pd.read_csv("../large_prime_numbers/test_results/Pollard_01-04-2025 22-54-43.csv")

# Группировка по 'number' и вычисление статистик:
stats2 = data2.groupby('number')['duration'].agg(
    mean_duration2='mean',           # среднее время
    median_duration2='median',       # медианное время
).reset_index()

stats2['number'] = stats2['number'].astype(float)

stats2['digits'] = stats2['number'].apply(lambda x: len(str(int(x))))
stats2['number_log10'] = stats2['number'].apply(lambda x: math.log10(x))

print(stats2)
```

```python
import matplotlib.pyplot as plt
# Настроим позиции по оси X для каждого уникального числа
#x_pos = range(len(stats))

# Создадим фигуру и ось
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 6))

# Отобразим медианные значения отдельными точками
ax1.scatter(stats['number_log10'], stats['median_duration'], color='red', zorder=5, label='Median')
ax2.scatter(stats2['number_log10'], stats2['median_duration2'], color='blue', zorder=5, label='Median')

ax1.set_xlabel('Log10 of number')
ax1.set_ylabel('Duration')
ax1.set_title('Time statistics for each number')
ax1.legend()

ax2.set_xlabel('Log10 of number')
ax2.set_ylabel('Duration')
ax2.set_title('Time statistics for each number')
ax2.legend()


plt.tight_layout()

plt.show()
```

```python

import pandas as pd

def read_test_result(filename):
    data = pd.read_csv(f"../large_prime_numbers/test_results/{filename}")

    digits = data["Digits"].to_list()
    N_values = list(map(int, data["N"].to_list()))
    factor_pollard = list(map(int, data["factor_pollard"].to_list()))
    duration_pollard = data["duration_pollard"].to_list()
    factor_trial = list(map(int, data["factor_trial"].to_list()))
    duration_trial = data["duration_trial"].to_list()
    factor_fermat = list(map(int, data["factor_fermat"].to_list()))
    duration_fermat = data["duration_fermat"].to_list()
    return digits, N_values, factor_pollard, duration_pollard, factor_trial, duration_trial, factor_fermat, duration_fermat
```

```python
import matplotlib.pyplot as plt
import numpy as np

#данные
digits, N_values, factor_pollard, duration_pollard, factor_trial, duration_trial, factor_fermat, duration_fermat = read_test_result("Pollard_Trial_Fermat_comparison_24-09-2024 19-21-26.csv")

#график зависимости времени от количества цифр
plt.figure(figsize=(10, 6))
plt.scatter(digits, duration_pollard, color='b', label='Pollard p-1')
plt.scatter(digits, duration_trial, color='r', label='Trial division')
plt.scatter(digits, duration_fermat, color='g', label='Fermat factorization')

plt.yscale('log')
plt.xlabel('Digits')
plt.ylabel('Duration (seconds)')
plt.title('Duration vs Number of Digits')
plt.grid(True)

z1 = np.polyfit(np.log(digits), np.log(duration_pollard), 1) 
p1 = np.poly1d(z1)
x_trend1 = np.linspace(min(digits), max(digits), 100)
y_trend1 = np.exp(p1(np.log(x_trend1)))
plt.plot(x_trend1, y_trend1, color='b', label='Trend Line Pollard p-1')

z2 = np.polyfit(np.log(digits), np.log(duration_trial), 1) 
p2 = np.poly1d(z2)
x_trend2 = np.linspace(min(digits), max(digits), 100)
y_trend2 = np.exp(p2(np.log(x_trend2)))
plt.plot(x_trend2, y_trend2, color='r', label='Trend Line Trial division')

z = np.polyfit(np.log(digits), np.log(duration_fermat), 1) 
p = np.poly1d(z)
x_trend = np.linspace(min(digits), max(digits), 100)
y_trend = np.exp(p(np.log(x_trend)))
plt.plot(x_trend, y_trend, color='g', label='Trend Line Fermat factorization')

plt.legend()
plt.show()

#график зависимости времени от N
plt.figure(figsize=(10, 6))
plt.scatter(N_values, duration_pollard, color='b', label='Pollard p-1')
plt.scatter(N_values, duration_trial, color='r', label='Trial division')
plt.scatter(N_values, duration_fermat, color='g', label='Fermat factorization')

plt.yscale('log')
plt.xscale('log')
plt.xlabel('N')
plt.ylabel('Duration (seconds)')
plt.title('Duration vs N')
plt.grid(True)

z1 = np.polyfit(np.log(N_values), np.log(duration_pollard), 1) 
p1 = np.poly1d(z1)
x_trend1 = np.linspace(min(N_values), max(N_values), 100)
y_trend1 = np.exp(p1(np.log(x_trend1)))
plt.plot(x_trend1, y_trend1, color='b', label='Trend Line Pollard p-1')

z2 = np.polyfit(np.log(N_values), np.log(duration_trial), 1) 
p2 = np.poly1d(z2)
x_trend2 = np.linspace(min(N_values), max(N_values), 100)
y_trend2 = np.exp(p2(np.log(x_trend2)))
plt.plot(x_trend2, y_trend2, color='r', label='Trend Line Trial division')

z = np.polyfit(np.log(N_values), np.log(duration_fermat), 1) 
p = np.poly1d(z)
x_trend = np.linspace(min(N_values), max(N_values), 100)
y_trend = np.exp(p(np.log(x_trend)))
plt.plot(x_trend, y_trend, color='g', label='Trend Line Fermat factorization')

plt.legend()
plt.show()

#график зависимости времени от количества цифр
plt.figure(figsize=(10, 6))
plt.scatter(factor_pollard, duration_pollard, color='b', label='Pollard p-1')
plt.scatter(factor_trial, duration_trial, color='r', label='Trial division')
plt.scatter(factor_fermat, duration_fermat, color='g', label='Fermat factorization')

plt.yscale('log')
plt.xscale('log')
plt.xlabel('Factor')
plt.ylabel('Duration (seconds)')
plt.title('Duration vs Factor')
plt.grid(True)


z1 = np.polyfit(np.log(factor_pollard), np.log(duration_pollard), 1) 
p1 = np.poly1d(z1)
x_trend1 = np.linspace(min(factor_pollard), max(factor_pollard), 100)
y_trend1 = np.exp(p1(np.log(x_trend1)))
plt.plot(x_trend1, y_trend1, color='b', label='Trend Line Pollard p-1')

z2 = np.polyfit(np.log(factor_trial), np.log(duration_trial), 1) 
p2 = np.poly1d(z2)
x_trend2 = np.linspace(min(factor_trial), max(factor_trial), 100)
y_trend2 = np.exp(p2(np.log(x_trend2)))
plt.plot(x_trend2, y_trend2, color='r', label='Trend Line Trial division')

z = np.polyfit(np.log(factor_fermat), np.log(duration_fermat), 1) 
p = np.poly1d(z)
x_trend = np.linspace(min(factor_fermat), max(factor_fermat), 100)
y_trend = np.exp(p(np.log(x_trend)))
plt.plot(x_trend, y_trend, color='g', label='Trend Line Fermat factorization')

plt.legend()
plt.show()
```
